```python
#Use scipy library to create a random uniform ditibution data (100000 numbers from 0 to 10)
#Plot the following density distribution. You can use fisize=(9,9), xlim=(-1,11)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Set the seed for reproducibility
np.random.seed(42)

# Generate the random uniform distribution
data = uniform.rvs(loc=0, scale=10, size=100000)

# Plot the density distribution
plt.figure(figsize=(9, 9))
plt.hist(data, density=True, bins=100, alpha=0.7)
plt.xlim(-1, 11)
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Random Uniform Distribution')
plt.show()
```

```python
#cdf (cumulative distribution function) gives you the area under the distribution's density curve
#to the left of a certain value on the x axis. Calculate cdf with x axis from 0 to 25.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Set the seed for reproducibility
np.random.seed(42)

# Define the random uniform distribution
rv = uniform(loc=0, scale=10)

# Define the range of x-axis values
x = np.linspace(0, 2.5, 100)

# Calculate the CDF
cdf = rv.cdf(x)

# Plot the CDF
plt.plot(x, cdf)
plt.xlabel('X')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function')
plt.grid(True)
plt.show()
```

```python
#ppf() is the inverse of cdf(): it returns the x axis cutoff value (quantile) associated with a
#given probability. Calculate the cutoff value for probability with 40%

import numpy as np
from scipy.stats import uniform

# Set the seed for reproducibility (optional)
np.random.seed(42)

# Define the random uniform distribution
rv = uniform(loc=0, scale=10)

# Define the desired probability
p = 0.4

# Calculate the cutoff value using PPF
cutoff_value = rv.ppf(p)

print("Cutoff value for probability", p, "is", cutoff_value)
```

```python
#Demonstrate for uniform distribution, x values within the range will have the same probability
#density. x values outside the range have a probability density of 0.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Set the seed for reproducibility
np.random.seed(42)

# Define the range for the uniform distribution
lower_limit = 2
upper_limit = 8

# Define the random uniform distribution within the range
rv = uniform(loc=lower_limit, scale=upper_limit-lower_limit)

# Generate random samples from the distribution
sample_size = 10000
samples = rv.rvs(size=sample_size)

# Plot the histogram of the samples
plt.hist(samples, density=True, bins=50, alpha=0.7)

# Plot the probability density function
x = np.linspace(lower_limit, upper_limit, 100)
plt.plot(x, rv.pdf(x), color='red', lw=2)

# Set plot labels and title
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution within Range')

# Set plot limits
plt.xlim(lower_limit-1, upper_limit+1)
plt.ylim(0, 1.2 * max(rv.pdf(x)))

# Show the plot
plt.show()
```

```python
#Use scipy's norm to have normal distribution. The normal distribution has default values of
#mean of 0 and standard deviation of 1. Investigate what percentage of data generated by a
#default normal distribution is below -1. What percentage of data generated is above 1 and
#percentage lies between -1 and 1. Print the three percentages.
#print the ppf for 2.5% cutoff and 97.5% cutoff.

from scipy.stats import norm

# Create the normal distribution object with default mean of 0 and standard deviation of 1
normal_dist = norm()

# Calculate the percentage of data below -1
percentage_below_minus1 = normal_dist.cdf(-1) * 100

# Calculate the percentage of data above 1
percentage_above_1 = (1 - normal_dist.cdf(1)) * 100

# Calculate the percentage of data between -1 and 1
percentage_between_minus1_and_1 = (normal_dist.cdf(1) - normal_dist.cdf(-1)) * 100

# Calculate the PPF for the 2.5% and 97.5% cutoffs
ppf_2_5_cutoff = normal_dist.ppf(0.025)
ppf_97_5_cutoff = normal_dist.ppf(0.975)

# Print the percentages and PPF values
print("Percentage of data below -1:", percentage_below_minus1)
print("Percentage of data above 1:", percentage_above_1)
print("Percentage of data between -1 and 1:", percentage_between_minus1_and_1)
print("ppf for 2.5% cutoff:", ppf_2_5_cutoff)
print("ppf for 97.5% cutoff:", ppf_97_5_cutoff)


```

```python
#We will use binomial distribution to model coin flips. The scipy name for the binomial is
#binom. Here we set the number of flips per trial to 10, the number of rials to 10000, and success
#probability to 0.5. Create a random binomial distribution data "fair_coin_flips" in the code
#line below

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Set the parameters
num_flips_per_trial = 10
num_trials = 10000
success_prob = 0.5

# Create the binomial distribution object
binomial_dist = binom(num_flips_per_trial, success_prob)

# Generate random samples from the binomial distribution
samples = binomial_dist.rvs(num_trials)

# Count the occurrences of each number of successful flips
unique_values, counts = np.unique(samples, return_counts=True)

# Plot the histogram of the samples
plt.bar(unique_values, counts/num_trials, alpha=0.7)

# Set plot labels and title
plt.xlabel('Number of Successful Flips')
plt.ylabel('Probability')
plt.title('Binomial Distribution')

# Show the plot
plt.show()
```

```python
#Continue from above binomial distribution(success rate=0.8), give the codes to get the probability
#of achieving 5 heads or less with 10 flips.

from scipy.stats import binom

# Set the parameters
num_flips_per_trial = 10
success_prob = 0.8
num_heads_or_less = 5

# Calculate the probability of achieving 5 heads or less
probability = binom.cdf(num_heads_or_less, num_flips_per_trial, success_prob)

# Print the probability
print("Probability of achieving 5 heads or less with 10 flips:", probability)


```

```python
#Continue from above binomial distribution(success rate=0.8), give the codes to get the probability
#of achieving 9 heads or more with 10 flips.

from scipy.stats import binom

# Set the parameters
num_flips_per_trial = 10
success_prob = 0.8

# Create the binomial distribution object
binomial_dist = binom(num_flips_per_trial, success_prob)

# Calculate the probability of getting 9 heads or more
probability = 1 - binomial_dist.cdf(8)

# Print the probability
print("Probability of achieving 9 heads or more with 10 flips:", probability)
```

```python
#For continuous probability density functions, you use pdf() to check the probability density at a
#given x value. For discrete distributions like the binomial, use stats.distribution.pmf()
#(probability mass function) to check the mass (proportion of observations) at given number
#of successes k Write codes to give the probability density at 5 heads with success rate
#of 0.5 of 10 flips.

from scipy.stats import binom

# Set the parameters
num_flips_per_trial = 10
success_prob = 0.5
num_heads = 5

# Create the binomial distribution object
binomial_dist = binom(num_flips_per_trial, success_prob)

# Calculate the probability density at 5 heads
probability_density = binomial_dist.pmf(num_heads)

# Print the probability density
print("Probability density at 5 heads with 10 flips:", probability_density)
```

```python
#What about the probability density at 8 heads with success rate of 0.8 of 10 flips?

from scipy.stats import binom

# Set the parameters
num_flips_per_trial = 10
success_prob = 0.8
num_heads = 8

# Calculate the probability density at 8 heads
probability_density = binom.pmf(num_heads, num_flips_per_trial, success_prob)

# Print the probability density
print("Probability density at 8 heads with 10 flips:", probability_density)
```
